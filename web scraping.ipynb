```python
from bs4 import BeautifulSoup
import requests

# Step 1: Make a request to the web page
url = 'https://scikit-learn.org/stable/modules/linear_model.html'  # Replace with the URL of the web page you want to scrape
response = requests.get(url)

# Step 2: Create a Beautiful Soup object
soup = BeautifulSoup(response.content, 'html.parser')

# Step 3: Extract the written text inside a particular tag
# Replace 'tag_name' with the specific HTML tag you want to extract text from, such as 'p', 'h1', 'div', etc.
# Here, we're using the 'p' tag as an example
paragraphs = soup.find_all('p')  # Find all the 'p' tags on the web page

# Extracted text list to store the extracted data
extracted_text_data = []

# Loop through each paragraph and extract the text
for paragraph in paragraphs:
    extracted_text_data.append(paragraph.get_text())  # Append the text inside each 'p' tag to the list

# Open a .txt file in write mode and write the extracted text into it
with open('extracted_data.txt', 'w', encoding='utf-8') as txtfile:
    for text in extracted_text_data:
        txtfile.write(text + '\n')
for paragraph in paragraphs:
    print(paragraph.get_text())
print('Data has been successfully saved to extracted_data.txt')
```
